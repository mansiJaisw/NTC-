```python
import matplotlib.pyplot as plt
import numpy as np

# draw plot as fig
fig, ax= plt.subplots()

x= np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),arrowprops=dict(facecolor='blue', shrink=0.04))

ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=0,angleB=-90"));

plt.show()

```

# ADDING ANNOTATAE & LEGEND

```python
import numpy as np
import matplotlib.pyplot as plt
# you can write  
# matplotlib import as plt


x= np.arange(0,10)
y= x^2
z= x^3

plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,z)

plt.title("matplotlib")
plt.xlabel("x axis")
plt.ylabel("y axis")




#Annotation

plt.annotate(xy=[2,1],text="first entery",rotation=25)
plt.annotate(xy=[4,7],text="secound entery",rotation=25)
plt.annotate(xy=[8,10],text="third entery",rotation=25)

#ADD LEGEND

plt.legend(["race 1","race 2","race 3"], loc = 4) # loc indect loction of box

plt.show()

```

# PIE PLOT

PIE CHART IS A CIRCULAR STASTICAL PLOT THAT CAN DISPLAY ONLY ONE SERIS

THE AREA OF THE CHART IS THE TOTAL % OF THE GIVEN DATA

```python
import numpy as np
import matplotlib.pyplot as plt

#creating dataset

branch = ["computer","civil","mechnical","it","electrical","automobile"]

data=[60,45,36,50,30,55]
explode = (0.1,0,0,0,0.1,0)

#creat chart

fig = plt.figure(figsize=(10,10))

plt.pie(data, labels=branch, explode=explode, autopct='%1.1f%%', shadow = True) #find pr value : autopct = '%1.1f%%'


plt.show()
```

# 3D PLOTING

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d  #3d tollkit lib
```

```python
fig = plt.figure()

#syntax for 3D 

ax = plt.axes(projection = '3d')

#define alll 3 axes

z= np.linspace(0,1,100)
x= z * np.sin(25*z)
y= z * np.cos(25*z)

#3d ploting

ax.plot3D(x,y,z,'r')
ax.set_title('3D ploting')
plt.show()
```
